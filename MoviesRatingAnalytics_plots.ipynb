```python
#Building Dashboards ...
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style("darkgrid")
f, axes = plt.subplots(2,2,figsize=(15,15))
k2 = sns.kdeplot(movies.Budget, movies.AudienceRating,\
                 ax=axes[0,0])
k3= sns.kdeplot(movies.Budget, movies.CriticsRating, \
                ax=axes[0,1])
k2.set(xlim=(-20, 160))
k3.set(xlim=(-20, 160))
z = sns.violinplot(data=movies, x='Genre', y='CriticsRating', \
                   ax=axes[1,0])
k1 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, shade=True, shade_lowest=False, cmap='Reds',\
                 ax=axes[1,1])
#tip
k1b = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, cmap='Reds', \
                  ax=axes[1,1])
plt.show()
```
