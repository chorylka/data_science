```python
#this is an example of a dashboard including some s
import pandas as pd
import os
movies = pd.read_csv("Movie-Ratings.csv")
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
movies.columns = ['Film','Genre', 'CriticsRating', 'AudienceRating',
       'Budget', 'Year']
```

```python
sns.set_style("dark", {"axes.facecolor": "black"}) #white, whitegrid, dark, darkgrid, ticks
f, axes = plt.subplots(2,2,figsize=(15,15))
#plot [0,0]
k2 = sns.kdeplot(movies.Budget, movies.AudienceRating,\
                 ax=axes[0,0], cmap='inferno', shade=True, shade_lowest=True)
k2b = sns.kdeplot(movies.Budget, movies.AudienceRating,\
                 cmap='cool', ax=axes[0,0])
#plot [0,1]
k3= sns.kdeplot(movies.Budget, movies.CriticsRating, \
                ax=axes[0,1], cmap='inferno', shade=True, shade_lowest=True)
k3b= sns.kdeplot(movies.Budget, movies.CriticsRating, \
                cmap='cool',ax=axes[0,1])
#plot [1,0]
z = sns.violinplot(data=movies, x='Genre', y='CriticsRating', \
                   ax=axes[1,0])
#plot [1,1]
k1 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, shade=True, shade_lowest=False, cmap='Reds',\
                 ax=axes[1,1])
#tip
k1b = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, cmap='gist_gray_r', \
                  ax=axes[1,1])
k2.set(xlim=(-20, 160))
k3.set(xlim=(-20, 160))

plt.show()
```
