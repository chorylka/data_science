```python
#here we import the csv file and create the dataframe
import pandas as pd
import os
movies = pd.read_csv("Movie-Ratings.csv")

#here we change some columns name to make it easier to read

movies.columns = ['Film','Genre', 'CriticsRating', 'AudienceRating',
       'Budget', 'Year']

#it's also required to change some columns to category

movies.Film= movies.Film.astype('category')
movies.Genre= movies.Genre.astype('category')
movies.Year= movies.Year.astype('category')


#here we prepare the plots library

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].Budget)
    mylabels.append(gen)
#print(list1)
sns.set_style("whitegrid") #this would make the grid more readable
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27) #this is a A4 paper size
h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.title("Movie Budget Distribution", fontsize=35, color="DarkBlue", fontname="Arial") #including the title in a good size font
plt.ylabel("Number of Movies", fontsize=25, color="Red")
plt.xlabel("Budget", fontsize=25, color="Red")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1,prop={'size':20})
plt.show()
```
