```python
import pandas as pd
import os
movies = pd.read_csv("Movie-Ratings.csv")
```

```python
movies
```

```python
movies.columns = ['Film','Genre', 'CriticsRatings', 'AudienceRatings',
       'Budget', 'Year']
```

```python
movies
```

```python
movies.Film= movies.Film.astype('category')
#transforma a coluna Films em category anteriormente era um objeto, em R isto se chama 'factors'
movies.Genre= movies.Genre.astype('category')
movies.Year= movies.Year.astype('category')
#transformando as colunas 'Genre' & 'Year' em category
#category é uma Finite list of text values :-)
```

```python
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].Budget)
    mylabels.append(gen)
#print(list1)
h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.legend()
plt.show()
```
