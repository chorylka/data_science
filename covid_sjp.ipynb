```python
import pandas as pd
data = pd.read_csv ("covid19-pr.csv")
#data.sample(100)
data_sjp = data.query('city == "São José dos Pinhais"')
data_sjp_final = data_sjp.sort_values('last_available_confirmed')
data_sjp_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
plt.rcParams['figure.figsize']=12,8
```

```python
data_sjp.info()
```

```python
sns.relplot(x="epidemiological_week", y="last_available_confirmed", data=data_sjp_final, kind="line");
```

```python
plt = sns.lineplot(x = data_sjp_final.date.astype("datetime64"), y= data_sjp_final.last_available_confirmed, data=data_sjp_final);
plt.set_xticklabels(data_sjp_final.date,rotation='45')
```
