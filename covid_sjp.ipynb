```python
import pandas as pd
data = pd.read_csv ("covid19-pr.csv")
#data.sample(100)
data_sjp = data.query('city == "São José dos Pinhais"')
data_sjp_final = data_sjp.sort_values('last_available_confirmed', ascending=False)
data_sjp_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
plt.rcParams['figure.figsize']=8,4
```

```python
data_sjp.info()
```

```python
sns.relplot(x="epidemiological_week", y="last_available_confirmed", data=data_sjp_final, kind="line");
```

```python
sns.relplot(x="date", y="last_available_confirmed", data=data_sjp_final, kind="line");
plt.xticks(list(range(0,20)), data_sjp_final.date, rotation='45') 
plt.gcf().set_size_inches(8, 8) # alterar tamanho
plt.show()
```
